```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.DataFrame({'kode': ['kode0', 'kode1', 'kode2', 'kode3'],
                     'JK': ['P', 'L', 'L', 'P'], # JK=Jenis Kelamin
                     'JB': ['3', '1', '2', '4']}) # JB=Jam Belajar(jam)/hari

data2 = pd.DataFrame({'kode': ['kode0', 'kode1', 'kode2', 'kode3'],
                      'Nilai EAS': ['94', '65', '87', '75'],
                      'Nilai': ['A', 'B', 'A', 'B']})

hasil = pd.merge(data1, data2, on='kode')
hasil
```

```python
data1 = pd.DataFrame({'kode': ['kode0', 'kode1', 'kode2', 'kode3'],
                      'Matkul': ['Matematika', 'Multivar', 'Multivar', 'MO'],
                     'JK': ['P', 'L', 'L', 'P'], # JK=Jenis Kelamin
                     'JB': ['3', '1', '2', '4']}) # JB=Jam Belajar(jam)/hari

data2 = pd.DataFrame({'kode': ['kode0', 'kode1', 'kode2', 'kode3'],
                      'Matkul': ['Teksam', 'MO', 'Multivar', 'ADW'],
                      'Nilai EAS': ['94', '65', '87', '75'],
                      'Nilai': ['A', 'B', 'A', 'B']})

hasil = pd.merge(data1, data2, on=['kode', 'Matkul'])
hasil
```

```python
hasil_left = pd.merge(data1, data2, how='left', on=['kode', 'Matkul'])
hasil_left
```

```python
hasil_outer = pd.merge(data1, data2, how='outer', on=['kode', 'Matkul'])
hasil_outer
```

```python
hasil_inner = pd.merge(data1, data2, how='inner', on=['kode', 'Matkul'])
hasil_inner
```

```python
data1 = pd.DataFrame({'JK': ['P', 'L', 'L'],
                     'JB': ['3', '1', '2']},
                    index=['kode0', 'kode1', 'kode2'])
data2 = pd.DataFrame({'Matkul': ['Teksam', 'MO', 'Multivar'],
                      'Nilai EAS': ['94', '65', '87']},
                    index=['kode0', 'kode1', 'kode3'])
hasil = data1.join(data2)
hasil
```

```python
hasil = data2.join(data1, how='outer')
hasil
```

```python
hasil = data2.join(data1, how='inner')
hasil
```

```python
hasil = pd.merge(data1, data2, left_index=True, right_index=True, how='outer')
hasil
```

```python
hasil = pd.merge(data1, data2, left_index=True, right_index=True, how='inner')
hasil
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [df1, df2, df3]
hasil = pd.concat(frames)
hasil
```

```python
hasil = pd.concat(frames, keys=['X', 'Y', 'Z'])
hasil
```
