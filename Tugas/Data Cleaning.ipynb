## Imputation Missing Value

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik
data['PLATING TANK'] = data['PLATING TANK'].fillna((data['PLATING TANK'].median())) 
data['PROOF CUT'] = data['PROOF CUT'].fillna((data['PROOF CUT'].mean()))
data['CALIPER'] = data['CALIPER'].fillna((data['CALIPER'].median()))
data.isnull().any()
```

```python
#imputasi untuk data numerik yang lain
data['INK TEMPERATURE']=data['INK TEMPERATURE'].fillna(31)
data['HUMIFITY']=data['HUMIFITY'].fillna(70)
data['ROUGHNESS']=data['ROUGHNESS'].fillna(5)
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data['PROOF ON CTD INK'] = data.fillna(data['PROOF ON CTD INK'].value_counts().index[0])
data['BLADE MFG'] = data.fillna(data['BLADE MFG'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Encode untuk Grain Screened, Proof On Ctd Ink, Direct Steam, Type On Cylinder tanpa membuat kolom baru
data['TYPE'] = LabelEncoder().fit_transform(data['CUSTOMER'])
data['GRAIN SCREENED'] = LabelEncoder().fit_transform(data['GRAIN SCREENED'])
data['PROOF ON CTD INK'] = LabelEncoder().fit_transform(data['PROOF ON CTD INK'])
data['DIRECT STEAM'] = LabelEncoder().fit_transform(data['DIRECT STEAM'])
data['TYPE ON CYLINDER'] = LabelEncoder().fit_transform(data['TYPE ON CYLINDER'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['WAX'] = StandardScaler().fit(data[['WAX']]).transform(data[['WAX']])
data['HARDENER'] = StandardScaler().fit(data[['HARDENER']]).transform(data[['HARDENER']])
data['ROLLER DUROMETER'] = StandardScaler().fit(data[['ROLLER DUROMETER']]).transform(data[['ROLLER DUROMETER']])
data['ANODE SPACE RATIO'] = StandardScaler().fit(data[['ANODE SPACE RATIO']]).transform(data[['ANODE SPACE RATIO']])
data.head()
```

## Normalize

```python
from sklearn.preprocessing import Normalizer
```

```python
data['WAX'] = Normalizer().fit(data[['WAX']]).transform(data[['WAX']])
data['HARDENER'] = Normalizer().fit(data[['HARDENER']]).transform(data[['HARDENER']])
data['ROLLER DUROMETER'] = Normalizer().fit(data[['ROLLER DUROMETER']]).transform(data[['ROLLER DUROMETER']])
data['ANODE SPACE RATIO'] = Normalizer().fit(data[['ANODE SPACE RATIO']]).transform(data[['ANODE SPACE RATIO']])
data.head()
```
